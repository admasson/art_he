```python
import numpy as np
import math
import os
import copy
import matplotlib.pyplot as plt

from astropy import constants as const # quickly access to all kind of physical constants
from scipy.interpolate import interpn, interp1d
from scipy.optimize import minimize
```

### Set parameters

We then set our model's parameters in a big dictionnary. Here we take the example of HD 209458 b

```python
params = {}

# planet & star properties
params['r_planet']           = 1.359 * const.R_jup.value / 1e3 # km (Bonomo2017)
params['r_star']             = 1.155 * const.R_sun.value / 1e3 # km (Bonomo2017)

# limb darkening coefficients: must have 0, 2, or 4 elements (0: no LD, 2: quadratic LD law, 4: non-linear LD law)
params['limb_dark_coeff']    = [0.212, 0.290] # (EXOFAST)

# orbital properties
params['a']                  = 0.04707 * const.au.value # semi-major axis (m) (Torres2008)
params['Porb']               = 3.52474955               # orbital period (days in BJD-TBD) (ExoClock)
params['e']                  = 0                        # eccentricity (fixed)
params['lbda']               = 0.1                      # spin-orbit orbital angle (°)
params['w']                  = 0                        # argument at periastron (°) (fixed) In case of circular orbit, w is -90 if define in the planetary orbit, and +90 if define in the stellar orbit (see pyasl.KeplerEllipse documentation)
params['i']                  = 86.71                    # orbital inclination (°) (Torres2008)
params['b']                  = np.cos(params['i']*np.pi/180.)*params['a']/(params['r_planet']/1e3) # impact parameter

# transit geometry
# transit phases (in time from mid transit: must be in days in BJD-TDB) at which computing the transit geometry
transit_duration             = 3.072 / 24 # days (Albrecht et al. 2012)
nb_transit_phase             = 10 # nb of transit phases for the sampling
params['time_samples']       = np.linspace(-transit_duration/2,+transit_duration/2,nb_transit_phase)
# size of the grid used for computing the transit geometry. The grid must have enough resolution such that the smallest altitude layers covers at least 2 pixels of the grid. 
# Otherwise, the model will throw an error and suggest a minimum nb of pixels for the transit grid size. If computation gets too long, another solution is to reduce the number of altitude layers or to space them linearly.
params['transit_grid_size']  = 2000

# atmospheric grid parameters
params['dnu']                = 1e-6 # wavelength step in the modeled spectrum (microns)
params['nrp']                = math.ceil(2*params['r_star']/params['r_planet']+1) # Vertical extension of the grid (in number of Rp) -> here we take the stellar diameter (in Rp) to ensure the absorption will cover the fulls tellar disk

# Broadening parameters
params['rad_vel']            = True # if True, applies the lat/longitudinal depedent velocities profile
params['fac_d']              = 3.67 # Doppler broadening (doppler thermal + turbulent, see Lampon et al 2020) -> corresponds to "alpha_D" has defined in Lampon et al 2020
params['vshift']             = 0    # km/s Overall shift of the whole signature (-> global escape velocity)

# Long/latitude dependent doppler shift. /!\ vb & vr must be >= 0 ! The appropriate sign will be given later in the program to get an appropriate blue/red shift
params['vb']                 = 0 # km/s >= 0, blueshift of the blueshifted part of the planet surface
params['vr']                 = 0 # km/s >= 0, redshift of the redshifted part of the planet surface
params['frac_red']           = 0 # fraction (from 0 to 1) of the planet surface corresponding to a redshift
params['frac_blue']          = 0 # fraction (from 0 to 1) of the planet surface corresponding to a blueshift

# whether wavelengths are defined in air or vacuum
params['air']                = False # if True, wavelengths are defined in air. False = vacuum

# verbose
params['verbose']            = True # If True then it will print the log
```

Let's parametrized the model:

```python
HD209_model.set_params(params)
```

We now load an abundance and velocity profile from the hydrodynamical code

```python
# paths to the grid
t_d = 10000    # Temperature, K
mlr = '2p37e11' # Mass loss rate, g/s in string format

abund_path = f'./Example_profiles/Abund_HD209_ratio90_T_{t_d}_MLR_{mlr}.txt'
vel_path   = f'./Example_profiles/Velocity_HD209_ratio90_T_{t_d}_MLR_{mlr}.txt'

abund_file = np.loadtxt(abund_path,skiprows=1) # 1st row is skipped because it contains text.
vel_file = np.loadtxt(vel_path,skiprows=1)

zlev = (abund_file[:,0]-1.)*params['r_planet'] # altitude in km relative to the "surface" ie the z1=1Rp level.

he3_profile = abund_file[:,5] # He3 density profile (atome/cm³)
vel_profile = vel_file[:,1] # km/s

plt.figure()
plt.plot(zlev/params['r_planet'],he3_profile,'k+-')
plt.xlabel('Altitude [Rp]')
plt.ylabel('He3 abundance')

plt.figure()
plt.plot(zlev/params['r_planet'],vel_profile,'k+-')
plt.xlabel('Altitude [Rp]')
plt.ylabel('He3 velocity [km/s]')
```

### Compute model for a given mass loss rate and temperature

Now that the grid is loaded and the model parametrised, we can compute the Helium triplet line using the profiles:

```python
# Compute the model
HD209_model.compute_model(zlev,t_d,he3_profile,vel_profile,plot=True)
```

### Convolve to CARMENES resolution & plot

The model can handle convolution with the isntrumental response of CARMENES (.convolve_CARMENES()) and SPIRou (.convolve_SPIRou()). Here we convolve to CARMENES resolution to get the final spectrum:

```python
HD209_model.convolve_CARMENES()
```

And plot the final spectrum

```python
# we can then plot the final spectrum along with the cumulative weight of each layer:
wavelengths = HD209_model.wno
helium_model = HD209_model.tro

plt.figure()
plt.plot(wavelengths,helium_model)
plt.xlabel('Microns')
plt.ylabel('Flux')
```

### Effective transit phase for fast computing

Note that we need to sample the transit geometry with a high-enough number of transit phases to avoid this parameter having an effect on the resulting line profile. From my estimations, we would need at least 100 transit phases to ensure the transit geometry is perfectly reproduced. This, of course, would make models computations waaaay too long for using it in retrievals. One solution is to use the "effective transit phase" technique from M.Lopez: finding a single transit phase (ie time from mid transit) whose signature alone reproduces the average of a hundred transit phases in the model. Below is an example on how to compute this effective transit phase and use it:

```python
# First we set a model with as few atmospheric layers as reasonnably possible to have a fast computation of a single transit phase. 
# For example, here we do a simple undersampling of the altitude grid and profiles
undersampling_factor = 3 # only take 1 out of 3 points in the profiles
he3_profile_undersample = he3_profile[::undersampling_factor]
vel_profile_undersample = vel_profile[::undersampling_factor]
zlev_undersample = zlev[::undersampling_factor]

# Then we set the 100 transit phases at which sampling the geometry:
params['time_samples'] = np.linspace(-transit_duration/2,+transit_duration/2,100)

# We update the model with this new parameter
HD209_100phaseModel = He_model()
HD209_100phaseModel.set_params(params)

# And we compute the model on the 100 phases but with a very low resolution over the altitude layers
HD209_100phaseModel.compute_model(zlev_undersample,t_d,he3_profile_undersample,vel_profile_undersample)
```

```python
# the above computation can take a few minutes (it took ~5min for me)
```

```python
# Once the 100-phase model has been computed, we use a chi² minimization to find the best time-from-mid transit whose model alone reproduces the 100-phase average
def chi2_func(time_from_mid):
    # compute the model at a single time from mid
    DummyModel = He_model(verbose=False)
    # we use the same parameters except for the mid transit
    copy_params = copy.deepcopy(params)
    copy_params['time_samples'] = np.array([time_from_mid])
    DummyModel.set_params(copy_params)
    # compute the model
    DummyModel.compute_model(zlev_undersample,t_d,he3_profile_undersample,vel_profile_undersample)
    # extract the fitting curve & model curve. Normalize by the mean of model curve to have values near unity
    y_model = np.copy(HD209_100phaseModel.trm)
    y_fit   = np.copy(DummyModel.trm)
    y_model /= np.nanmean(y_model)
    y_fit   /= np.nanmean(y_model)
    # and return the reduced chi² between the DummyModel & the 100-phase model
    chi2 = np.nansum( (y_fit - y_model)**2 )
    print(f'\rTime from mid: {time_from_mid[0]} d, chi2 = {chi2}',end='',flush=True)
    return chi2

# Now minimize the above to find the effective transit time. Our first guess x0 is set at mid-transit, we set bounds such as planet is transiting
res = minimize(chi2_func,x0=0., bounds=[(-1*transit_duration/2,+1*transit_duration/2)])

print(f'\nDone, found effective time from mid transit to be: {res.x} d')
```

```python
# Verify by comparing the effective-phase model & the 100-phase averaged model
effective_time_from_mid = res.x
# initialize
EffPhase_Model = He_model()
# update parameters
EffPhase_params = copy.deepcopy(params)
EffPhase_params['time_samples'] = np.array([effective_time_from_mid])
EffPhase_Model.set_params(EffPhase_params)
# compute
EffPhase_Model.compute_model(zlev_undersample,t_d,he3_profile_undersample,vel_profile_undersample)
```

```python
# Plot for comparison
plt.figure()
plt.plot(EffPhase_Model.wn,EffPhase_Model.trm,label='Effective phase')
plt.plot(HD209_100phaseModel.wn,HD209_100phaseModel.trm,label='100-phase averaged Model')
plt.xlabel('microns')
plt.ylabel('flux')
plt.legend()
```

Now that we found the effective time from mid transit, we can use it to compute the model with more altitude layers and in a faster way. This allows to compute $\chi^2$ map in a reasonnable amount of time.

```python
# update params
fast_params = copy.deepcopy(params)

# update the time samples using the effective phase
fast_params['time_samples']       = np.array([effective_time_from_mid])

# size of the grid used for computing the transit geometry. The grid must have enough resolution such that the smallest altitude layers covers at least 2 pixels of the grid. 
# Otherwise, the model will throw an error and suggest a minimum nb of pixels for the transit grid size. If computation gets too long, another solution is to reduce the number of altitude layers or to space them linearly.
fast_params['transit_grid_size']  = 2000

# initialize
HD209_FastModel = He_model()
# parametrize
HD209_FastModel.set_params(fast_params)
# compute
HD209_FastModel.compute_model(zlev,t_d,he3_profile,vel_profile,plot=True)
# convolve
EffPhase_Model.convolve_CARMENES()
```

```python
# The above model is computed in ~14 sec instead of >1min before. We plot it here:
plt.figure()
plt.plot(HD209_FastModel.wn,HD209_FastModel.trm,label='Effective phase Model')
plt.xlabel('microns')
plt.ylabel('flux')
plt.legend()
```
